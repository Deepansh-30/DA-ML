## Q.4

-Plot Stacked Bar Graph for the above data.
    - Use proper colors and x and y label for the graph
    - Use legend positioned at upper right corner
    - Also mark xticks with an angle of 90 Degree

```python
store=['Delhi','Mumbai','Chennai','Kolkata','Hyderabad']
Clothing=np.array([82,78,48,46,39])
Equipment=np.array([48,31,51,29,18])
Accessories=np.array([15,20,14,13,9])
ind=[z for z, _ in enumerate(store)]
plt.bar(ind,Clothing,width=0.5,label='Clothing',color='#0d31b1',bottom=Equipment+Accessories)
plt.bar(ind,Equipment,width=0.5,label='Equipments',color='#ffa600',bottom=Accessories)
plt.bar(ind,Accessories,width=0.5,label='Accessories',color='#fd2361')
plt.xticks(ind,store,rotation=90,Horizontalalignment="right")
plt.xlabel("Metropolis(city)" ,color="#003f5c")
plt.ylabel("Stores",color="#003f5c")
plt.legend(loc='upper right',fancybox=True, framealpha=0.8, shadow=True, borderpad=1,title='stores in cities')
plt.title("Shops among various Metropolis")
```

## Q.5 For Titanic Dataset
- Load Dataset using Seaborn
- Plot a Bar Graph of Survival of Passenger against the Passenger Class.
- Use proper "hue" attribute, and palette color. Mark 'x' and 'y' label with proper title.
- Use legend located at upper left corner

```python
print(sns.get_dataset_names(),end=" ")
```

```python
data=sns.load_dataset('titanic')
data.head(6)
```

```python
survive=sns.countplot(x='pclass',hue='survived',palette='Set2',data=data)
survive.set(title='Survival of passenger against the Passenger class')
plt.legend(loc="upper left",fancybox=True, framealpha=0.8, shadow=True, borderpad=1,title='survival vs pclass')
plt.show()
```

## Q.9 


- Plot the Pie Chart for the two colums with some explode for the country "India"  and proper shadow

```python
Country='US','Great Britain','India','Russia','Germany'
Medal=np.array([46,27,26,19,17])
colors=['#45af31','#dbe240','#ff5e3b','#3564af','#a20ee8']
explode=(0,0,0,0,0)
plt.pie(Medal,explode=explode,labels=Country,colors=colors,shadow=True,autopct='%1.2f%%',startangle=90, wedgeprops={"edgecolor":"cyan",'linewidth': 1,
                    'linestyle': 'dashed', 'antialiased': True})
plt.legend( Country,title="Acheivement",loc="best")
plt.axis('equal')
plt.tight_layout()
plt.show()
```
